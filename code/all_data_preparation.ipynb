```python
import pandas as pd

# pulling in the dataframs to be merged
df_movie_genre_long = pd.read_csv('../data/movie_genre_long.csv', index_col=0)
df_tmdb = pd.read_csv('../data/tmdb_filtered.csv', index_col=0)

# filtering and renaming the tmdb columns
df_tmdb_filtered = df_tmdb[['budget', 'revenue', 'imdb_id','vote_average','vote_count']]
df_tmdb_filtered.rename(columns={'budget':'budget_api', 'revenue':'revenue_api'}, inplace=True)

# merging the data to make it harder, better, faster, stronger
df_all_data = df_movie_genre_long.merge(df_tmdb_filtered,
                         how='inner',
                         left_on='movie_id',
                         right_on='imdb_id')

df_all_data.head(10)
```
