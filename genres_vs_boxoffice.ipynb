#### df_all_data I will allow me to look at the individual genres and how they performed against box office and profitability

```python
# take a look at the columns
df_all_data.info()
```

#### I'll add a profitability column that will show the difference between worldwide gross and production budget.

NOTE: Microsoft is a large enough company where they may want to distribute large movies worldwide. This is why I'll gauge profitability based off (worldwide_gross - production_budget) rather than (domestic_gross - production_budget)

```python
# add profitability column to df_all_data
df_all_data['profitability'] = df_all_data['worldwide_gross'] - df_all_data['production_budget']
```

```python
# take a look at the new column
df_all_data.head()
```

#### We don't need the person's name, ratings, and etc.. Only columns we need are ['movie_id', 'genre', 'primary_title', 'production_budget', 'domestic_gross', 'worldwide_gross']

#### Now I need to group by the dataframe in order to get the genres and gross associated with the movie_id. Then from there I can do a final group by to find the average of each genre.

```python
# we only need the relevant columns for our analysis
relevant_columns = ['movie_id', 
                    'genre', 
                    'primary_title', 
                    'production_budget', 
                    'domestic_gross', 
                    'worldwide_gross', 
                    'profitability']
df_all_filtered = df_all_data[relevant_columns].groupby(['movie_id','genre']).mean().reset_index()
```
